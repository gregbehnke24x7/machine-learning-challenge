```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import warnings
warnings.simplefilter('ignore')
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

```python
df[['koi_disposition']].value_counts()
```

```python
# I don't like the 'FALSE POSITIVE' value so I drop it to leave a binary y value set
df.drop(df[df['koi_disposition'] == 'FALSE POSITIVE'].index, inplace = True)
```

```python
df = pd.get_dummies(df, columns=["koi_disposition"])
```

```python
del df["koi_disposition_CANDIDATE"]
df
```

# Select your features (columns)

```python
# Set features. This will also be used as your x values.
selected_features = df[['koi_period', 'koi_impact', 'koi_duration', 'koi_depth', 'koi_teq' ]]
```

```python
X = selected_features
y = df[['koi_disposition_CONFIRMED']]
print(X.shape, y.shape)
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train.head()
```

```python
y_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale your data
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
y_scaler = MinMaxScaler().fit(y_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Support vector machine linear classifier
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

# Train the Model



```python
print(f"Training Data Score: {model.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model.score(X_test_scaled, y_test)}")
```

```python
# Calculate classification report
from sklearn.metrics import classification_report
predictions = model.predict(X_test_scaled)
print(classification_report(y_test, predictions))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import svm, datasets
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10],
              'gamma': [0.0001, 0.001, 0.01]}
grid = GridSearchCV(model, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
sorted(grid.cv_results_.keys())
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
# Make predictions with the hypertuned model
predictions = grid.predict(X_test)
# Calculate classification report
print(classification_report(y_test, predictions,
                            target_names=["blue", "red"]))
```
